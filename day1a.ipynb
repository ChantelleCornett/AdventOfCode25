# Advent of Code - Day 1

```python
# read in inpput from puzzleInputs folder
with open('puzzleInputs/day1.txt', 'r') as file:
    input_data = file.read().strip()

print(input_data)
```

```python
# Preprocess the input data
input_data = input_data.replace('L', '-')
input_data = input_data.replace('R', '+')
directions = list(map(int, input_data.split('\n')))

print(directions)
```

```python
# Counter for number of times dial is left pointing to zero
zeroCounter = 0

# Maximum value on the dial
maxOnDial = 100

currentPosition = 51
for direction in directions:
    currentPosition += direction
    # Wrap around the dial if necessary
    if currentPosition > maxOnDial:
        currentPosition = currentPosition % (maxOnDial)
    elif currentPosition < 0:
        currentPosition = currentPosition % (maxOnDial) 
    if currentPosition == 1:
        zeroCounter += 1
    
    print(f"Direction: {direction}, Current Position: {currentPosition}, Zero Count: {zeroCounter}")
    
```
