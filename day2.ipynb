```python
input_data = "11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124"
```

```python
range_list = []

range_list = input_data.split(',')
for i in range(len(range_list)):
    range_list[i] = range_list[i].split('-')


print(range_list)
```

```python
pattern = re.compile(r'^(?!0)(\d+)\1$')
invalid = []
for range_low_high in range_list:
    low = int(range_low_high[0])
    high = int(range_low_high[1])

    for i in range(low, high + 1,1):
        str_i = str(i)
        if pattern.search(str_i):
            invalid.append(i)

```

```python
print(sum(invalid))
```
